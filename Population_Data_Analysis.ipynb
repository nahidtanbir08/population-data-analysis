```python
import pandas as pd

df = pd.read_csv('population.csv')
```

# 1.  Mask array for Total Population > 6K

```python
mask = df['Total population'] > 6000


print(mask.values)
```

# 2. Min and Max Registered Voters

```python
min_voters = df['registered voters'].min()
max_voters = df['registered voters'].max()

print(f"Min Registered Voters: {min_voters}")
print(f"Max Registered Voters: {max_voters}")
```

# 4. Find the District and registered voters for the highest "total population"

```python
highest_pop_idx = df['Total population'].idxmax()
district_name = df.loc[highest_pop_idx, 'District']
voters_count = df.loc[highest_pop_idx, 'registered voters']

print(f"District: {district_name}")
print(f"Registered Voters: {voters_count}")
```

# 5. Find the average population for the entire Nepal 

```python
avg_population = df['Total population'].mean()

print(f"Average Population: {avg_population}")
```

# 6. Find the first 15 districts information about voters and total population

```python
first_15_districts = df[['District', 'registered voters', 'Total population']].head(15)

print(first_15_districts)
```
